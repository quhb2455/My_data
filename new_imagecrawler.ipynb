```python
from selenium import webdriver
import urllib.request
from time import sleep

#url 경로로 다운로드
def downImage(url, name):
    dir = 'C:\\Users\\Sub\\PycharmProjects\\test_img'
    urllib.request.urlretrieve(url, dir + name + '.jpg')

search_term = '박보영'
url = "https://www.google.com/search?q="+ search_term +"&tbm=isch"

browser = webdriver.Chrome("C:\\Users\Sub\Desktop\\aboutPython\chromedriver_win32\chromedriver.exe")

browser.get(url)

pic_name = browser.find_elements_by_class_name("rg_i")

flg = 0
img_url = {}


#url 경로 모음.
for idx, value in enumerate(browser.find_elements_by_class_name("rg_i")):
    print(idx,"번째")
    #value.screenshot(search_term + "_" + str(idx) + ".png")
    if value.get_attribute("data-src") == None :
        img_url[idx] = value.get_attribute("src")

    else:
        img_url[idx] = value.get_attribute("data-src")

    #print(value.get_attribute("data-src"))
    #screen_data[idx] = value.get_attribute("data-src")
    print(img_url[idx])
    downImage(img_url[idx],'박보영_'+str(idx))

    if flg == 49:
        break
    flg += 1


browser.close()

```
