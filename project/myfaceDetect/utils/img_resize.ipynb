```python
import os,sys
import cv2

original_path = '../face_img/myface'

imglists = os.listdir(original_path)

for imglist in imglists :
    img = cv2.imread(original_path+'/'+imglist)
    print(imglist)
    resizing_img = cv2.resize(img, dsize=(224,224), interpolation = cv2.INTER_AREA)
    cv2.imwrite("../face_img/resize/"+imglist, resizing_img)
    
    
```
