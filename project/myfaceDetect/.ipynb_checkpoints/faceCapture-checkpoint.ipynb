```python
import os,sys
import cv2

original_path = '../face_img/b_face2'
imglists = os.listdir(original_path)
cascade_file = r"C:\Users\Sub\Anaconda3\envs\sub\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml"
flag = 1

for imglist in imglists :
    img = cv2.imread(original_path+'/'+imglist)
    print(imglist)
    img_gs = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cascade = cv2.CascadeClassifier(cascade_file)
    face_list = cascade.detectMultiScale(img_gs, scaleFactor = 1.1,  minNeighbors = 4, minSize = (70,70))
    
    print(flag)
    flag += 1
    overlap_naming = 1
    
    for face in face_list :
        x,y,w,h = face
        print(face)
        resizing_img = cv2.resize(img[y:y+h, x:x+w], dsize=(200,200), interpolation = cv2.INTER_AREA)
        cv2.imwrite("../face_img/myface/"+str(overlap_naming)+"_"+imglist, resizing_img)
        overlap_naming += 1
    
```
